#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#code
def SD(data, ddof=0):
    n = len(data)
    mean = sum(data)/n
    return np.sqrt(sum((x - mean) ** 2 for x in data)/ (n - ddof))
````

````python
#set up the sat and act observations below 

sat_samples = merged['2017_sat_participation']
act_samples =merged['2017_act_participation']
````

**STEP 1: Null and Alt hypothese**
   
$H_0$: The true mean difference in participation rate between the SAT and ACT participation rates is 0

$H_A$: The true mean difference in participation rate between the SAT and ACT participation rates is not 0
    
**STEP 2: Level of significance**
$\alpha = 0.05$
 
**STEP 3: Test Stat**


````python
# Code:
import scipy.stats as stats
from statsmodels.stats import weightstats as stests

sat_mean = sat_samples.mean()
act_mean = act_samples.mean()

sat_std = sat_samples.std()
act_std = act_samples.std()


ztest ,p_val = stests.ztest(sat_samples, x2=act_samples, value=0,alternative='two-sided')
p_val
````

````python
if p_val <0.05:
    print('reject null hypothesis')
else: 
    print('accept null hypothesis')
````
